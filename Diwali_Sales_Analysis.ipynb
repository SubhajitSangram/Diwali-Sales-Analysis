## EDA- Exploratory data analysis

```python
ax=sns.countplot(data=df,x='Gender',color='g'or 'r')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show(ax)
```

```python
plt.figure(figsize=(5,2.5))
slaes_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=slaes_gen,x='Gender',y='Amount',dodge=False,palette={'M': 'Red', 'F': 'green'})
plt.show()
```

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
plt.figure(figsize=(5,2.5))
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
plt.show()
```

```python

sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).tail(10)

sns.barplot(x = 'State',y= 'Orders' ,data = sales_state)
plt.show()
```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
plt.show()
```

```python
ax=sns.countplot(data=df,x='Marital_Status')
plt.show()
```

```python
Sales_Status=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,2.5)})
sns.barplot(data = Sales_Status, x = 'Marital_Status',y= 'Amount', hue='Gender')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
plt.show()
```

```python
sales_catagory = df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_catagory, x = 'Product_ID',y= 'Amount')
plt.show()
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
plt.show()
```

## Key Findings

#### Female customers made more purchases and generated higher total sales than male customers.
#### Customers in the 26–35 age group contributed the highest number of purchases and the largest sales amount
#### The top-performing states generated a significant share of total sales, indicating strong regional demand.
#### Married customers, especially married women, spent more during the Diwali shopping season.
#### Customers working in sectors such as IT, Healthcare, and Aviation were among the highest spenders.
#### Certain product categories consistently achieved the highest sales, making them the most popular among customers.
